```python
import pandas as pd
import os
file = "first"
num = "_up&down"
acceler_csv_file_name = 'Accelerometer'
gyro_csv_file_name = 'Gyroscope'
pressure_csv_file_name = 'Pressure'

start_time = 20
end_time = 140

new_dir_path = f'{file}{num}_filter'

try:
    os.mkdir(new_dir_path )
except:
    pass

try:
    # CSVファイルを読み込む
    df = pd.read_csv(f'{file}/{acceler_csv_file_name}.csv')

    df = df[(df['t'] >= start_time) & (df['t'] <= end_time)]

    # 書き換えたデータをCSVファイルに保存する
    df.to_csv(f'{new_dir_path}/{acceler_csv_file_name}.csv', index=False)
except:
    print(f"{acceler_csv_file_name}.csv is not found")
    pass
try:
    # CSVファイルを読み込む
    df = pd.read_csv(f'{file}/{gyro_csv_file_name}.csv')
    
    df = df[(df['t'] >= start_time) & (df['t'] <= end_time)]

    # 書き換えたデータをCSVファイルに保存する
    df.to_csv(f'{new_dir_path}/{gyro_csv_file_name}.csv', index=False)
except:
    print(f"{gyro_csv_file_name}.csv is not found")
    pass
try:
    # CSVファイルを読み込む
    df = pd.read_csv(f'{file}/{pressure_csv_file_name}.csv')

    df = df[(df['t'] >= start_time) & (df['t'] <= end_time)]

    # 書き換えたデータをCSVファイルに保存する
    df.to_csv(f'{new_dir_path}/{pressure_csv_file_name}.csv', index=False)
except:
    print(f"{pressure_csv_file_name}.csv is not found")
    pass
```
