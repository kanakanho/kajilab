```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
```

```python
"""
y = (TxPower - 10) * n * log10(x) + baius
逆関数
x = 10 ** ((y - baius) / (TxPower - 10) * n)
0 <= x <= 25
"""
```

```python
TxPower = 0
n = 0.2
baius = -5
```

```python
y_s = []
y2_s = []

x_lim_minus = 0
x_lim_plus = 25
for x in np.arange(x_lim_minus, x_lim_plus, 0.01):
    y = ((TxPower - 10) * n * np.log10(x) + baius )
    y2 = -((TxPower - 10) * n * np.log10(x) - baius)
    y_s.append(y)
    y2_s.append(y2)
fig = plt.figure(figsize=(10, 10))
plt.title('受信電力と距離の関係の陰関数')
# plt.plot(np.arange(x_lim_minus, x_lim_plus, 0.01), y_s)
plt.plot(np.arange(x_lim_minus, x_lim_plus, 0.01), y2_s)
plt.gca().set_aspect('equal', adjustable='box')
plt.xlabel('距離[m]')
plt.ylabel('受信電力[dBm]')
plt.grid()
plt.show()
```

```python
x_s = []
y_lim_minus = -8
y_lim_plus = -2
for y in np.arange(y_lim_minus, y_lim_plus, 0.01):
    x = 10 ** ((y - baius) / ((TxPower - 10) * n))
    x_s.append(x)
fig = plt.figure(figsize=(20, 20))
plt.plot(x_s, np.arange(y_lim_minus, y_lim_plus, 0.01))
plt.gca().set_aspect('equal', adjustable='box')
plt.title('電波を妨害する要素がない場合のRSSIと距離の関係')
plt.xlabel('距離[m]')
plt.ylabel('受信電力[dBm]')
plt.grid()
plt.show()
```
